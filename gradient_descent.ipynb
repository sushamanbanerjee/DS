<b>Gradient Descent</b> is an optimization algorithm used to minimize the cost function in machine learning and deep learning models. Here's a brief overview:

<b>Objective:</b> Find the values of parameters (weights, biases) that minimize the error in prediction by adjusting them iteratively.

<b>Working:</b>

<ul>It starts with an initial guess for the parameters.</ul>
<ul>In each iteration, the algorithm calculates the gradient (partial derivative) of the cost function concerning each parameter.</ul>
<ul>Parameters are updated in the opposite direction of the gradient to reduce the cost function.</ul>
<ul>Learning Rate: This determines the step size for updating parameters. A small learning rate results in slow convergence, while a large learning rate may overshoot the minimum.</ul>

<b>Types:</b>

Batch Gradient Descent: Uses the entire dataset to compute gradients.

Stochastic Gradient Descent (SGD): Uses one data point at a time.

Mini-Batch Gradient Descent: Uses a small random subset of the dataset.

Convergence: The algorithm continues until the gradient is very small or reaches a predefined number of iterations.

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_derivatives(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return df_dx, df_dy

def gradient_descent(x, y, learning_rate=0.1, epochs=100):
    graph=[(x,y)]
    cost_val=[]
    for epoch in range(epochs):
        df_dx, df_dy = compute_derivatives(x, y)
        x = x - learning_rate * df_dx
        y = y - learning_rate * df_dy
        graph.append((x,y))
        if epoch % 10 == 0:
            cost = x**2 + y**2
            cost_val.append(cost)
            print(f'Epoch {epoch}: x = {x}, y = {y}, Cost = {cost}')
    
    return x, y, graph, cost_val
```

```python
X = np.array([1])
y = np.array([3])
m, c, graph, cost = gradient_descent(X, y, learning_rate=0.01, epochs=1000)
print(f'Final parameters: m = {m}, c = {c}')
```

```python
x_vals = np.linspace(-4, 4, 400)
y_vals = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2 + Y**2
plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20),cmap="jet")
plt.plot([p[0] for p in graph], [p[1] for p in graph], 'ro-', label="Gradient Descent Path")
plt.title('Gradient Descent on $f(x, y) = x^2 + y^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot([p for p in cost], 'ro-')
plt.title("Cost Function Graph")
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.show()
```
